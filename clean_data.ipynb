```python
import pandas 

df = pandas.read_excel("data/data_dropout_59-64.xlsx")

# remove last Row have NaN
df = df.drop(labels=[4120],axis=0)
df
```

```python
df.columns
```

```python
df = df.drop(df[df['เกรดปี1เทอม1'].isna() | df['เกรดปี1เทอม2'].isna() | df['GPA_SCHOOL'].isna()].index)
df = df.drop(df[df['GPA_SCHOOL'] > 4].index)
df['GPA_SCHOOL'].unique()
```

```python
data_test = df.groupby("ADMIT_YEAR").sample(n=10, random_state=10)
df = df.drop(data_test.index)
data_test
```

```python
from pycaret.classification import setup, compare_models, predict_model
```

```python
DropOut_Feature = ['ADMIT_YEAR', 'ADMIT_TERM', 'DEPT_ID', 'GPA_SCHOOL']

# STUDY_STATUS
experiment = setup(df, target="STUDY_STATUS", categorical_features=DropOut_Feature)
```

```python
best_model = compare_models()
```

```python
predict_model(best_model, data_test)
```
