```python
import pandas as pd
from pycaret.classification import *
import numpy as np
from sklearn.metrics import confusion_matrix
```

# LOAD_DATA AND COMPAREMODELS

```python
pd.set_option('display.max_columns',None)
```

```python
clean_data = pd.read_csv('../data/data_dropout_59-64(final_model).csv')
```

```python
test = clean_data.sample(n = int(849*0.3),random_state=10)
```

```python
test.to_csv('../data/data_dropout_59-64(final_model_test).csv',index=False)
```

```python
train = clean_data.drop(test.index)
```

```python
train.to_csv('../data/data_dropout_59-64(final_model_train).csv',index=False)
```

```python
s = setup(train,target='STUDY_STATUS')
```

```python
models = compare_models()
```

```python
models
```

```python
predictions = predict_model(models,data=test)
```

```python
con = confusion_matrix(y_true=predictions['STUDY_STATUS'],y_pred=predictions['prediction_label'],labels=['R','G'])
```

```python
con
```

```python
save_model(models,'model_top_no_STILL_STUDENT_2Class')
```
